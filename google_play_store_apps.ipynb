# Finding the null values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
unique_values = df['Category'].unique()
```

```python
unique_values
```

```python
df.describe()
```

```python
# Create a mapping dictionary of unique Category values to corresponding Rating values
mapping = df.groupby('Category')['Rating'].mean().to_dict()

# Fill null values in Rating column using the mapping dictionary
df['Rating'] = df['Rating'].fillna(df['Category'].map(mapping))
```

```python
df["Rating"]
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

# # **Data Visualiation**

```python
import matplotlib.pyplot as plt
```

# **Q1: - Which is the most used Android version?**

```python
# Count the occurrences of each Android version
android_version_counts = df['Android Ver'].value_counts()

# Get the most used Android version
most_used_android_version = android_version_counts.index[0]
most_used_android_version
```

```python
# Plot the graph
plt.figure(figsize=(15, 6))
android_version_counts.plot(kind='bar')
android_version_counts.plot(kind='bar', color='skyblue')
# Add labels and title
plt.xlabel('Android Version')
plt.ylabel('Count')
plt.title('Count of Android Versions')

# Display the graph
plt.show()
```

# **Q2: - Which all categories of mob apps are existing from this dataset?**

```python
# Get the unique categories of mobile apps
categories = df['Category'].unique()

print(categories)
```

# **Q3: - Which is the most installed category of mob app from this dataset?**

```python
# Calculate the count of unique app values for each category
category_counts = df.groupby('Category')['App'].nunique()

# Find the category with the highest count
most_installed_category = category_counts.idxmax()

print(most_installed_category)
```

```python
# Plot the horizontal bar chart with bounding box
plt.figure(figsize=(15, 10))
plt.barh(category_counts.index, category_counts.values, color='orange')
plt.box(on=True)

# Add labels and title
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Count of Unique Apps by Category')

# Display the chart
plt.show()
```

# **Q4: - How many total categories are there?**

```python
# Calculate the total number of categories
total_categories = df['Category'].nunique()
total_categories
```

# **Q5: - Which apps have got 5-star ratings**

```python
# Filter the DataFrame for apps with 5-star ratings
five_star_apps = df[df['Rating'] == 5.0]['App']
# Calculate the total count of 5-star rated apps
total_count = len(five_star_apps)

print(five_star_apps)
```

***There are total 274 apps got 5 star rating***

# **Q6: - Which are the top 10 mob apps based on ratings?**

```python
# Group the DataFrame by app and calculate the sum of ratings
app_ratings_sum = df.groupby('App')['Rating'].sum()

# Sort the values in descending order and select the top 10 apps
top_10_apps = app_ratings_sum.nlargest(10)

# Plot the bar graph with colors and a bounding box
colors = ['skyblue', 'orange', 'green', 'purple', 'yellow', 'pink', 'blue', 'red', 'cyan', 'magenta']
top_10_apps.plot(kind='bar', color=colors)
plt.box(on=True)

# Add labels and title
plt.xlabel('App')
plt.ylabel('Sum of Ratings')
plt.title('Top 10 Mobile Apps based on Sum of Ratings')

# Display the graph
plt.show()
```
